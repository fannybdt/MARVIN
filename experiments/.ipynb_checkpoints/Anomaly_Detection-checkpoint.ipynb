```python
from marvin import MARVIN
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import ListedColormap, BoundaryNorm
from sklearn.manifold import TSNE
from sklearn.metrics import f1_score, balanced_accuracy_score
import torch
from torch.nn import functional as F
from torch.utils.data import DataLoader, Dataset
M = 8
K = 11
D = 16
mymodel = MARVIN(M, K, D, 8, 0.2, 2)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
mymodel.load_state_dict(torch.load("wandb_follow/MARVIN_CHU_normal_final.pt", weights_only = True))
mymodel.to(device)
```

```python
"""
cluster_labels = {
    0: 'BLASTES CD34 (not LyBlastes)',
    1: 'DEBRIS',
    2: 'DOUBLETS',
    3: 'GRAN IMMATURE (PROMYELO)',
    4: 'GRAN INTERMEDIATE MATURATION',
    5: 'LB',
    6: 'LT_NK',
    7: 'LyBlastes Type I',
    8: 'LyBlastes Type II',
    9: 'LyBlastes Type III',
    10: 'MONOCYTES',
    11: 'MRD',
    12: 'NEUTRO MATURES',
    13: 'PLASMOCYTES'
}
"""
cluster_labels = {
    0: 'BLASTES CD34 (not LyBlastes)',
    1: 'GRAN IMMATURE (PROMYELO)',
    2: 'GRAN INTERMEDIATE MATURATION',
    3: 'LB',
    4: 'LT_NK',
    5: 'LyBlastes Type I',
    6: 'LyBlastes Type II',
    7: 'LyBlastes Type III',
    8: 'MONOCYTES',
    9: 'MRD',
    10: 'NEUTRO MATURES',
    11: 'PLASMOCYTES'
}
cluster_labels_pred =  {
    0: 'BLASTES CD34 (not LyBlastes)',
    1: 'GRAN IMMATURE (PROMYELO)',
    2: 'GRAN INTERMEDIATE MATURATION',
    3: 'LB',
    4: 'LT_NK',
    5: 'LyBlastes Type I',
    6: 'LyBlastes Type II',
    7: 'LyBlastes Type III',
    8: 'MONOCYTES',
    9: 'NEUTRO MATURES',
    10: 'PLASMOCYTES'
}
sns.set_style("darkgrid")
colors = sns.color_palette('tab20', n_colors=len(cluster_labels)) 
tick_positions = [i + 0.5 for i in range(len(cluster_labels))]
bounds = range(len(cluster_labels) + 1)
lighter_colors = ListedColormap([(np.array(c) * 0.6 + 0.4) for c in colors])
norm = BoundaryNorm(bounds, len(colors)) 
    
```

```python
root = './CHU/MO_markers_logicle.csv'
df = pd.read_csv(root)
```

```python
import pandas as pd

c = pd.factorize(df['label'], sort = True)[0] 
x = df.iloc[:, :M]  
marker_names = x.columns[:-1]

x_all = torch.tensor(x.values, dtype=torch.float32) 
c_all = torch.tensor(c, dtype=torch.long) 

allset = torch.utils.data.TensorDataset(x_all, c_all)

dataloader = DataLoader(allset, 4000, shuffle=True)
```

```python
x, c_true = next(iter(dataloader))
x = x.to(mymodel.device)
c_true = c_true.to(mymodel.device)

q_c_x = F.softmax(mymodel.q_c_x(x), dim=-1)
c = torch.multinomial(q_c_x, 1, replacement=True).squeeze() 
```

```python
import umap.umap_ as umap

codes = mymodel.one_hot(c, mymodel.K)
q_z_cx = mymodel.q_z_cx([codes, x])
mu_z = q_z_cx[:, :mymodel.D]
logvar_z = q_z_cx[:, mymodel.D:]
z = mu_z + logvar_z.exp().sqrt() * torch.randn_like(mu_z)

reducer = umap.UMAP(n_components=2, min_dist=0.5, n_neighbors=8)
zu = reducer.fit_transform(z.detach().cpu().numpy())
plt.figure(figsize=(8, 6))
scatter = plt.scatter(
    zu[:, 0],
    zu[:, 1],
    c=[list(cluster_labels.keys()).index(label) for label in c_true.cpu().numpy()],
    cmap=ListedColormap(colors),
    norm=norm,
    s=15,
    alpha=0.7
)

colorbar = plt.colorbar(scatter, ticks=np.arange(len(cluster_labels)), boundaries=bounds)
colorbar.set_ticks(tick_positions)
colorbar.ax.set_yticklabels(cluster_labels.values())
plt.title(r"UMAP of $q(z|c, x)$")
plt.xlabel("UMAP Component 1")
plt.ylabel("UMAP Component 2")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig("exp1_umap.png", dpi=150)
plt.show()
plt.close()
```

```python
p_x_z = mymodel.p_x_z(z)
x_recon = p_x_z[:, :M] + p_x_z[:, M:].exp() ** 0.5 * torch.randn(len(x), M, device = device)
```

```python
import umap.umap_ as umap
import matplotlib.pyplot as plt
import matplotlib as mp
sns.set_style("darkgrid")


reducer = umap.UMAP(n_components=2, min_dist=0.5, n_neighbors = 8)
embedding_original = reducer.fit_transform(x.cpu().numpy())
embedding_recon = reducer.transform(x_recon.detach().cpu().numpy())

plt.figure(figsize=(8, 6))

scatter_original = plt.scatter(
    embedding_original[:, 0],
    embedding_original[:, 1],
    c=[list(cluster_labels.keys()).index(label) for label in c_true.cpu().numpy()],
    cmap=lighter_colors,
    norm=norm,
    s=15,
    alpha=0.6,  
    label="Original"
)

scatter_recon = plt.scatter(
    embedding_recon[:, 0],
    embedding_recon[:, 1],
    c=[list(cluster_labels.keys()).index(label) for label in c_true.cpu().numpy()],
    cmap=ListedColormap(colors),
    norm=norm,
    s=15,
    alpha=0.8,
    label="Reconstruction"
)

colorbar = plt.colorbar(scatter_recon, ticks=np.arange(len(cluster_labels)), boundaries = bounds)
colorbar.set_ticks(tick_positions)
colorbar.ax.set_yticklabels(cluster_labels.values())

plt.title(r"UMAP: Original vs Reconstruction $p(x|z)$")
plt.xlabel("UMAP Component 1")
plt.ylabel("UMAP Component 2")
plt.grid(alpha=0.3)
plt.legend(loc = 'best')
plt.tight_layout()
plt.savefig("umap_reconstruction.png", dpi=150)
plt.show()
plt.close()
```

```python
reconstruction_losses = []
for i in range(x.shape[0]):
    predicted_class = c[i].item()
    true_class = c_true[i].item()
    codes = mymodel.one_hot(torch.tensor([predicted_class]), mymodel.K).to(mymodel.device)
    q_z_cx = mymodel.q_z_cx([codes, x[i].unsqueeze(0)])
    mu_z = q_z_cx[:, :mymodel.D]
    logvar_z = q_z_cx[:, mymodel.D:]
    z = mu_z + logvar_z.exp().sqrt() * torch.randn_like(mu_z)
    
    x_recon = mymodel.p_x_z(z)
    mu_x = x_recon[:, :mymodel.M]
    logvar_x = x_recon[:, mymodel.M:]
    loss = (0.5 * ((x[i] - mu_x) ** 2 / logvar_x.exp() + logvar_x)).sum(dim=-1)
    reconstruction_losses.append((true_class, loss.item()))

reconstruction_df = pd.DataFrame(reconstruction_losses, columns=['Class', 'Loss'])

plt.figure(figsize=(12, 6))
for class_label, class_name in cluster_labels.items():
    class_data = reconstruction_df[reconstruction_df['Class'] == class_label]['Loss']
    class_mean = class_data.mean()
    label_with_mean = f"{class_name} (μ={class_mean:.2f})"
    sns.kdeplot(
        class_data,
        label=label_with_mean,
        fill=True,
        color=colors[class_label]
    )
plt.xlim((-40, 60))
plt.xlabel(r'-log p(x|z) = $\dfrac{1}{2} \left( \dfrac{(x - \mu_x)^2}{\sigma_x^2} + \log \sigma_x^2 \right)$')
plt.ylabel('Density')
plt.title('Negative Log-Likelihood of Reconstruction per True Class')
plt.legend()
plt.savefig('reconstruction_loss_all.pdf', dpi=200)
plt.show()
```
